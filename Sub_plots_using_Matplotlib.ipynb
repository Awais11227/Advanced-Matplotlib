# Subplots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
batters = pd.read_csv('c:/Users/hp/Downloads/batter.csv')
batters.head()
```

```python
plt.figure(figsize=(11,5))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Batters')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,6))

ax.scatter(batters['avg'],batters['strike_rate'],color='green',marker='+')
ax.set_title('Batters')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'],color='black')

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'],color='green')
ax[1,0].hist(batters['avg'],color='black')
ax[1,1].hist(batters['runs'])
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])
```
