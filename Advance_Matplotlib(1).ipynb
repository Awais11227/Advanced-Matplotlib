```python
# pie -> single and multiple
df = pd.DataFrame(
    {
        'batsman':['Babar ','Rizwan','Saim','Afridi','Haris','Abdullah'],
        'match1':[120,90,35,45,12,10],
        'match2':[0,1,123,130,34,45],
        'match3':[50,24,145,45,10,90]
    }
)

df.head()
```

```python
df['match1'].plot(kind='pie',labels=df['batsman'].values,autopct='%0.1f%%')
```

# Multiple charts

```python
# multiple pie charts

df[['match1','match2','match3']].plot(kind='pie',subplots=True,figsize=(15,8))
```

```python
# multiple separate graphs together
# using stocks

stocks.plot(kind='line',subplots=True)
```

```python
# Multiindes dataframs
tips.pivot_table(index='day', columns='sex',values='total_bill',aggfunc='mean').plot(kind='line')
```

```python
# Multiindes dataframs
tips.pivot_table(index=['day','time'], columns=['sex','smoker'],values='total_bill',aggfunc='mean').plot(kind='line')
```

```python
# Multiindes dataframs
tips.pivot_table(index=['day','time'], columns=['sex','smoker'],values='total_bill',aggfunc='mean').plot(kind='bar')
```

```python
tips.pivot_table(index=['day','time'],columns=['sex','smoker'],values='total_bill',aggfunc='mean').plot(kind='pie',subplots=True,figsize=(20,10))
```
